# ndarray 연산

## element-wise 연산
- 기본적으로 두 배열은 동일한 shape여야 한다.
- 같은 위치 (인덱스)의 요소끼리 연산이 수행 된다.
- 결과 배열 또한 동일한 shape를 가진다.

```python
import numpy as np

arr1 = np.arange(1,10).reshape(3,3)
print(arr1)

arr2= np.full_like(arr1,2)
print(arr2)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(arr1 // arr2)
print(arr1 % arr2)
print(arr1 ** arr2)
```

## broadcasting 연산 방식
- 두 배열의 shape이 다르면 broadcasting 규칙에 따라 연산이 수행 된다.
- 차원이 작은 배열을 자동으로 확장하여 연산이 가능하도록 한다.

### 규칙
- 뒤쪽(마지막 차원)부터 shape를 비교
- 두 차원이 같거나, 둘 중 하나가 1이면 연산 가능
- 둘 다 다르고 둘 다 1이 아니면 broadcasting 불가

```python
# 브로드캐스트 연산 테스트
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.full((3,),2)

print(arr1, arr1.shape)
print(arr2, arr2.shape)

# (2,3) 배열과 (3,) 배열의 연산
# (3,) -> (1,3) --> (2,3)
print(arr1 + arr2)
```

```python
arr3 = np.array([1,2])
print(arr3, arr3.shape)
# ValueError: operands could not be broadcast together with shapes (2,3) (2,) 
# print(arr1 + arr3)
```

```python
# arr1과 다양한 연산
print(arr1 + 10)       # (2,3) + () 스칼라도 브로드캐스팅 적용
print(arr1 + [10])     # (2,3) + (1,) 브로드캐스팅 적용
print(arr1 + [[10]])   # (2,3) + (1,1) 브로드캐스팅 적용
```

NumPy에서는 다음 세 가지 방법으로 행렬곱을 수행할 수 있다.
- @ 연산자 -> 가장 기본적으로 사용
- np.matmul()

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(a@b)
print(np.matmul(a,b))
print(np.dot(a,b))
```

```python
c = np.array([[1,2]])       # (1, 2)
d = np.array([[3], [4]])    # (2, 1)
print(c.shape, d.shape)
x = c @ d
print(x.shape, x)
```

```python
c = np.array([[1,2,3], [4,5,6]])    
d = np.array([[7,8], [9,10], [11,12]])     
print(c.shape, d.shape)
x = c @ d
print(x.shape, x)
```

```python
c = np.array([[1,2,3], [4,5,6]])  
d = np.array([[7,8,9], [10,11,12]])   
print(c.shape, d.shape)
# x = c @ d                     3과 2는 행렬 곱 불가 ValueError 발생
# print(x.shape, x)
```

## 기타 연산
Numpy는 배열 전체에 대해 다양한 수학 함수를 제공한다.

```python
# 절대값
arr1 = np.array([[1,-2], [-3,4]])
print(np.abs(arr1))
```

```python
# 반올림
arr2 = np.array([-1.234, 1.234, 2.345, 3.456, 4.567])
print(np.round(arr2))         # 반올림
print(np.ceil(arr2))          # 올림
print(np.floor(arr2))         # 내림
print(np.trunc(arr2))         # 절삭
```

### axis 속성
연산을 수행할 축을 지정한다.

- axis=0 : 행 방향으로 연산 (열 기준 연산)
- axis=1 : 열 방향으로 연산 (행 기준 연산)
- axis=None : 배열을 1차원으로 펼친 후 연산

axis는 "어느 방향으로 줄일것인가"를 의미

```python
# min
arr = np.array([[1,2,3], [4,5,6]])
print(np.min(arr))
print(np.min())    # aaxis=None 이 기본 값으로 배열을 1차원으로 펼치고 최소값을 구함

print(arr.min(axis=0))
print(arr.min(axis=1))
```

```python
# sum
print(arr.sum())
print(arr.sum(axis=0))
print(arr.sum(axis=1))
```

```python
# mean
print(arr.mean())
print(arr.mean(axis=0))
print(arr.mean(axis=1))
```
