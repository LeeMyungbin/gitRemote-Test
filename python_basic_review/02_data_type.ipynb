# 숫자형(Numeric)
-정수(int)
-실수(float)

```python
# 사칙연산
n1=2
n2=3
print(n1+n2)
print(n1-n2)
print(n1*n2)
print(n1/n2)
print(n1%n2)   #나누기 후 나머지 반환
print(n1//n2)  #나누기 후 몫 반환
print(n1**n2)  #제곱
```

# 논리형(Boolean)
-True, False
-""으로 감싸면 X
-첫 글자 반드시 대문자
*연쇄비교 가능


```python
print( 1 == 3 )  # 같은지 
print( 1 != 3 )  # 같지 않은지
age=23
print(23< age <= 25)  #연쇄 비교
```

# 문자열(string)
-문자열 * 숫자  : 문자열 숫자만큼 반복
-replace(old,new) : 문자열을 치환하는 메서드
-strip[chars] : 문자 제거
-upper(): 대문자로 변경
-lower(): 소문자로 변경
-capitalize(): 첫 글자만 대문자로 변경
-swapcase() : 대문자는 소문자로, 소문자는 대문자로 변경
-title(): 단어의 첫글자를 대문자로 


```python
name="이명빈"
nickname="맹구리"
print("안녕하세요 제 이름은 " + name + "입니다. 별명은 " + nickname +"입니다")
print(name * 3)
date="2026/06/30"
rep_date=date.replace("/","-")
print("제 생일은 ", rep_date , "입니다.")

myself="hello, my name is milie"
print(myself.strip())              #양쪽 공백 제거
print(myself.strip("h"))           #양쪽 h 제거
print(myself.lstrip("hell"))       #선행 제거
print(myself.rstrip("lie"))        #후행 제거
```
