```python
stored_id = "admin"
stored_pw = "pass123"

while True:            #일치할때까지 반복 
    u_id = input("아이디: ")    #입력을 받을 수 있는 함수 
    u_pw = input("비밀번호: ")
    
    if u_id == stored_id and u_pw == stored_pw:
        print("환영합니다, 관리자님!")
        break
    else:
        print("아이디 또는 비밀번호가 틀렸습니다. 다시 시도하세요.")
```

```python
for i in range(1, 31):
    s = str(i)    #문자열 변환 
    if ('3' in s) or ('6' in s) or ('9' in s):   
        print("짝!", end=" ")    #위치인자, 키워드 (공백)
    else:
        print(i, end=" ")
```

```python
cart = {"Laptop": 1000000, "Mouse": 15000, "Monitor": 200000, "Pad": 5000}
discounted_cart = {k: v * 0.9 for k, v in cart.items() if v >= 50000}    #key와 value 값
print(discounted_cart)
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i}*{j}={i*j}", end="\t") # 탭 공백으로 가로 정렬
    print() # 한 단이 끝나면 줄바꿈
```

```python
visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]

# 언패킹과 zip을 활용한 분리
dates, counts = zip(*visitor_data)  # * : 언팩킹 연산자 

# 평균 계산
avg_visitors = sum(counts) / len(counts)

print(f"날짜: {dates}")
print(f"방문자 평균: {avg_visitors}명"
```
