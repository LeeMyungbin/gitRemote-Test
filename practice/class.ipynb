```python
class OrderItem:
    def __init__(self,name,unit_price,qty):
        self.name= name
        self.unit_price=unit_price
        self.qty=qty

    def total_price(self):
        return self.unit_price * self.qty
    
    def __str__(self):
        return f'{self.name} x{self.qty} = {self.total_price()}원'
        

item1 = OrderItem("아메리카노", 3000, 2)
item2 = OrderItem("치즈케이크", 5500, 1)

print(item1)
print(item2)

total = item1.total_price() + item2.total_price()
print(f'총합: {total}원')


# 기대 출력 예시
# 아메리카노 x2 = 6000원
# 치즈케이크 x1 = 5500원
# 총합: 11500원
```

```python
class MenuItem:
    def __init__(self, name, unit_price):
        self.name = name
        self.unit_price = unit_price
    
    def get_info(self):
        pass

class Coffee(MenuItem):
    def __init__(self, name, unit_price, shot=1):
        super().__init__(name, unit_price)    #부모 생성자 호출
        self.shot = shot
    
    def get_info(self):
        return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"

class Dessert(MenuItem):
    def __init__(self, name, unit_price, size="M"):
        super().__init__(name, unit_price)
        self.size = size
        
    def get_info(self):
        return f"{self.name}({self.size}) - {self.unit_price}원"

items = [
    Coffee("아메리카노", 3000, 2),
    Dessert("치즈케이크", 5500, "L"),
]

for item in items:
    print(item.get_info())
```

```python
class Cart:
    def __init__(self):
        self.items = []

    def add(self, item):
        self.items.append(item)

    def total(self):
        total_sum = 0
        for item in self.items:
            total_sum += item.total_price()
        return total_sum

    def receipt(self):
        for item in self.items:
            print(item) 
        print(f"총 금액: {self.total()}원")

# --- 테스트 코드 ---
# 1. 물건 생성 (OrderItem 객체들)
order1 = Coffee("아메리카노", 3000, 2)
order2 = Dessert("치즈케이크", 5500, "L")

# 2. 카트 생성 및 담기
my_cart = Cart()
my_cart.add(order1)
my_cart.add(order2)

# 3. 영수증 출력
my_cart.receipt()
```
