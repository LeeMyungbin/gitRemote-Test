#캡슐화 
-private 속성(변수)을 선언해서 외부에서 직접 접근을 막고, 제공 된 메소드를 통해서 사용하도록 강제한다.
-주요한 데이터가 외부에서 노출 되지 않는 은닉한 효과를 얻을 수 있다.

```python
class Account:
     def __init__(self, username, balance):
        # private 접근 제한 처리
        self.__username =username
        self.__balance = balance
     def deposit(self,money):
         """입금 메소드
          -money 유효성 검사
          -오염 된 입금 처리 방지
         """
         if money <= 0:
             print("입금액이 유효하지 않습니다.")
         else: 
           self.__balance +=money
           print("입금이 정상처리 되었습니다")
     def withdwaw(self,money):
         if money <=0:
             print("출금액이 유효하지 않습니다.")
         elif money > self._balance:
             print("잔액이 부족합니다.")
         else:
             self.__balance -= money
             print("출금이 정상처리 되었습니다.")
     def info(self):
        print(f"{self.__username}님의 계좌의 현재 잔액은 {self.__balance}원입니다.")


        

#파이선에서 private는 약속이다
#_클래스명_변수명으로 이름이 바뀐 것이므로 접근하려고 하면 접근 가능하지만 개발자 사이의 약속으로 외부 접근 x
acc = Account('hong', 1_000_000)
print(acc.username)
print(acc.balance)
#권장하지 않는 방법
print(acc._Account__username)
print(acc._Account__balance)
```

-하지만 개발자 사이의 약속으로 외부에서 접근하지 않도록 한다.

##왜 캡슐화를 할까?
